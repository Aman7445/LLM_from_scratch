```python
import gensim.downloader as api
model = api.load("word2vec-google-news-300")
```

# Example of a word as a vector

```python
word_vectors=model

#Lets look at how embeddings of a word looks like
print(word_vectors['son']) # Example: Accessing the vector for the word 'son'
```

```python
print(word_vectors['cat'].shape)
```

# Similar words

# King + Woman - Man = ?

```python
# Example of using most similar
print(word_vectors.most_similar(positive=['king','woman'], negative=['man'],topn=10))
```

# Let us check the similarity b/w a few pair of words

```python
print(word_vectors.similarity('woman','man'))
print(word_vectors.similarity('son','daughter'))
print(word_vectors.similarity('land','nephew'))
```

# Now let us see the vector similarity

```python
import numpy as np
#Words to compare
word1='man'
word2='woman'

word3='semiconductor'
word4='earthworm'

#Calculate the vector differences
vector_diff1=model[word1]-model[word2]
vector_diff2=model[word3]-model[word4]

#calculate the magnitude of vector difference
magnitude_of_difference1 = np.linalg.norm(vector_diff1)
magnitude_of_difference2 = np.linalg.norm(vector_diff2)
# Print the magnitude of the difference
print("The magnitude of the difference between '{}' and '{}' is {:.2f}".format(word1, word2, magnitude_of_difference1))
print("The magnitude of the difference between '{}' and '{}' is {:.2f}".format(word3, word4, magnitude_of_difference2))
```
